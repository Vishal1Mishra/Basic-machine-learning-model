```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.head()
```

```python
df_train.shape
```

```python
train_X = df_train.drop(columns=['Item_Outlet_Sales','Outlet_Establishment_Year'],axis=1)
train_y = df_train['Item_Outlet_Sales']
```

```python
train_X.head()
```

```python
test_X = df_test.drop(columns=['Item_Outlet_Sales','Outlet_Establishment_Year'],axis=1)
test_y = df_test['Item_Outlet_Sales']
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_X,train_y)
```

```python
preds_y = lr.predict(test_X)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(test_y,preds_y)
```

```python
filename = 'finalized_model.sav'
pickle.dump(lr, open(filename, 'wb'))
```
